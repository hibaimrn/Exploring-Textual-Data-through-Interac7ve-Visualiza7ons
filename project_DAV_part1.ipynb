```python
from convokit import Corpus, download
import pandas as pd

# Download the movie corpus and create a Corpus object
corpus = Corpus(filename=download("movie-corpus"))


```

```python
# Convert the corpus data to a list of dictionaries
corpus_data = []
for convo in corpus.iter_conversations():
    for utt in convo.iter_utterances():
        corpus_data.append({
            'conversation_id': convo.id,
            'utterance_id': utt.id,
            'speaker': utt.speaker.id,
            'text': utt.text,
            # Add other attributes as needed
        })

# Create a DataFrame using the list of dictionaries
df = pd.DataFrame(corpus_data)

# Specify the file path where you want to save the CSV file
csv_file_path = "movie_corpus.csv"

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)
```
